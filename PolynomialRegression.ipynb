```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures #for polynomial regression
import matplotlib.pyplot as plt
```

```python
demand=[5,10,15,20,27,35,45,65,95,150,200]
price=[5,10,15,20,25,30,35,40,45,50,55]
df=pd.DataFrame({"demand":demand,
                "price":price})
```

```python
df
```

```python
plt.scatter(df["demand"],df["price"],s=100,color='blue',edgecolors='red')
plt.title('Demand&Price_Elasticity')
plt.xlabel('Demand')
plt.ylabel('Price')
plt.grid('True')
plt.show()
```

Graphic shows that data performed polynomial so we will apply polynomial regression.

But first we are applying linear reggression for see difference between polynomial and linear regression.

```python
x=df['demand']
y=df['price']
```

```python
x=sm.add_constant(x)
model=sm.OLS(y,x).fit()
y_pred=model.predict(x)
```

```python
plt.scatter(df["demand"],df["price"],s=100,color='blue',edgecolors='red')
plt.plot(x,y_pred)
plt.title('Demand&Price_Elasticity')
plt.xlabel('Demand')
plt.ylabel('Price')
plt.grid('True')
plt.show()
```

### Polynomial Regression

$ y=b_{0}+b_{1}x+b_{2}x^2+b_{3}+x^3   \newline$
$b_{0}\rightarrow are\ bias $

```python
x_poly=x.to_numpy()
y_poly=y.to_numpy()

poly_reg = PolynomialFeatures(degree = 5) #degree is arbitrary
X_poly = poly_reg.fit_transform(x_poly)

#model education
pol_reg = LinearRegression()
pol_reg.fit(X_poly, y_poly)


```

```python
y_pred_poly=pol_reg.predict(poly_reg.fit_transform(x_poly))
```

```python
plt.scatter(df["demand"],df["price"],s=100,color='blue',edgecolors='red')
plt.plot(x,y_pred_poly)
plt.plot(x,y_pred)
plt.title('Demand&Price_Elasticity')
plt.xlabel('Demand')
plt.ylabel('Price')
plt.grid('True')
plt.show()
```

If we increase to degree polynomial regresson line will be more fit.
